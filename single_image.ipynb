```python
test_img_path = ds_path + "\Test_0.jpg"
test_img_path
```

```python
main_img = cv2.imread(test_img_path)
img = cv2.cvtColor(main_img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
plt.imshow(gray,cmap='Greys_r')
```

```python
gray.shape
```

```python
blurred = cv2.GaussianBlur(gray, (35,35),0)
plt.imshow(blurred,cmap='Greys_r')
```

```python
ret_otsu,im_bw_otsu = cv2.threshold(blurred,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
plt.imshow(im_bw_otsu,cmap='Greys')
```

```python
kernel = np.ones((30,30),np.uint8)
closing = cv2.morphologyEx(im_bw_otsu, cv2.MORPH_CLOSE, kernel)
```

```python
plt.imshow(closing,cmap='Greys')
```
